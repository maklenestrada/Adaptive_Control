### Discussion 

The performance of the least-squares approach was better than the composite update law and the baseline update law. The tracking error for the least-squares approach was the first to settle to zero at approximately 7.5 seconds. The composite approach also nearly settles to zero with some slight damping still occuring. The baseline approach was still oscillating near zero at 30 seconds while the other two approaches had settled at zero. All of the approaches matches the analysis where we would eventually see $e \rightarrow 0$ as $t \rightarrow \infty$.

For the parameter estimation, the least-squares approach was significantly better. The L2-norm of the parameter estimation error was zero during the entire simulation, while the composite and baseline approaches eventually settled to approximately 0.01 and 0.05, respectively. This matches the analysis, as the baseline approach is the simplest and only accounts for the tracking error, so we would expect the worst performance. The composite approach adds a prediction error, which improves parameter estimation by using both the tracking error and prediction error. The best performance is seen with the least-squares approach, which has a normalization factor for the prediction error and a forgetting factor that allows the $\Gamma$ matrix to adapt during the simulation.
