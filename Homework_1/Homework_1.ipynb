<div style="text-align: center; font-weight: bold;"> Part 2 </div>

Show that the controller 
$$
\dot{y} = x^{2}, \;\; u = -yN(y)x
$$
where N(y) is a Nussbaum gain function, results in boundedness of all signals and convergence of the output x and the internal state z to zero regardless of the values of a, c, A, B, and b as long as $b \neq 0$

\begin{align*}
\text{Dynamics: } \dot{x} &= ax + bu + c^{T}z \\
\dot{z} &= Az + Bx\\
\text{Controller: } \dot{y} &= x^{2} \\
u &= -y N(y)x\\
\text{Closed Loop: } \dot{\eta} &= 
\begin{bmatrix} \dot{x} \\ \dot{z} \\ \dot{y}     \end{bmatrix}  \\
&= \begin{bmatrix} ax - byN(y)x + c^{T}z \\  Az + Bx  \\x^2    \end{bmatrix} \\
&= f(\eta)\\
\text{Lyapunov Function: } V(\eta) &= \frac{1}{2}x^2 + \frac{1}{2}z^{T}Pz\\
\text{Khalil Theorem 4.6 pg136 : } P = P^{T}>0&\text{ so that } A^{T}P+PA= -Q \text{ with } Q>0\\
\text{Derrivative of Lyapunov Function: } \frac{d}{dt}V(\eta) &= x\dot{x} + z^{T}P\dot{z} \\
&= x\left( ax - byN(y)x + c^{T}z \right) + z^{T}P\left(Az + Bx    \right)\\
&=  ax^{2} - byN(y)x^{2} + c^{T}zx+ z^{T}PAz + xz^{T}PB\\
&=  \left(a - byN(y)\right) \dot{y} + c^{T}zx+ z^{T}PAz + xz^{T}PB\\
&z^{T}PAz \text{   term}\\
z^{T}PAz &= (z^{T}PAz)^{T}\\
(z^{T}PAz)^{T} &= z^{T}A^{T}P^{T}z\\
&= z^{T}A^{T}Pz\\
\longrightarrow z^{T}PAz &= \frac{1}{2}\left( z^{T}PAz +  z^{T}A^{T}Pz  \right)\\
&= \frac{1}{2}z^{T}\left(PA + A^{T}P\right)z\\
&= -\frac{1}{2}z^{T} Q z\\
\frac{d}{dt}V(\eta) &= \left(a - byN(y)\right) \dot{y} + c^{T}zx-\frac{1}{2}z^{T} Q z+ xz^{T}PB\\
&= \left(a - byN(y)\right) \dot{y} -\frac{1}{2}z^{T}Qz + c^{T}zx + xz^{T}PB\\
&= \left(a - byN(y)\right) \dot{y} -\frac{1}{2}z^{T}Qz + x \left(c + PB\right)^{T}z \\
&(c + PB)^{T} \text{ is a constant } \beta \\
&= \left(a - byN(y)\right) \dot{y} -\frac{1}{2}z^{T}Qz + x \beta z \\
&\text{where } \| x\beta z\| \leq \|x\| \| \beta \| | z \| \\
&\leq \left(a - byN(y)\right) \dot{y} -\frac{1}{2}z^{T}Qz + \|x\| \| \beta \| | z \| \\
&\text{Apply Youngs Inequality to }\|x\| \| \beta \| | z \| \\
&\|x\| \| \beta \| | z \| \leq \frac{\lambda^2}{2}\|x\|^2 + \frac{\beta}{2 \lambda^2} \|z \|^2\\
&\leq \left(a - byN(y)\right) \dot{y} -\frac{1}{2}z^{T}Qz+ \frac{\lambda^2}{2}\|x\|^2 + \frac{\beta}{2 \lambda^2} \|z \|^2\\
&\leq \left(a - byN(y)\right)\|x\|^2 -\frac{1}{2}z^{T}Qz+ \frac{\lambda^2}{2}\|x\|^2 + \frac{\beta}{2 \lambda^2} \|z \|^2\\
&\leq \left(\frac{\lambda^2}{2} + a - byN(y)\right)\|x\|^2 -\frac{1}{2}z^{T}Qz + \frac{\beta}{2 \lambda^2} \|z \|^2\\
&\leq \left(\frac{\lambda^2}{2} + a - byN(y)\right)\|x\|^2 -\frac{1}{2}\|Q\| \|z\|^2 + \frac{\beta}{2 \lambda^2} \|z \|^2\\
&\leq \left(\frac{\lambda^2}{2} + a - byN(y)\right)\|x\|^2 + \left(\frac{\beta}{2 \lambda^2}-\frac{1}{2}\lambda_{min}(Q) \right)\|z \|^2\\
&\leq \left(\frac{\lambda^2}{2} + a - byN(y)\right)\|x\|^2 - \left(\frac{1}{2}\lambda_{min}(Q) - \frac{\beta}{2 \lambda^2} \right)\|z \|^2\\
&\text{We choose } \frac{1}{2}\lambda_{min}(Q) - \frac{\beta}{2 \lambda^2}  \text{ such that it is } > 0\\ 
&\leq -\left(byN(y) - \frac{\lambda^2}{2} - a\right)\|x\|^2 - \left(\frac{1}{2}\lambda_{min}(Q) - \frac{\beta}{2 \lambda^2} \right)\|z \|^2\\
&\text{We choose }byN(y) - \frac{\lambda^2}{2} - a \text{ such that it is } > 0\\ 
\longrightarrow \dot{V} &\leq -\left(byN(y) - \frac{\lambda^2}{2} - a\right)\|x\|^2 - \left(\frac{1}{2}\lambda_{min}(Q) - \frac{\beta}{2 \lambda^2} \right)\|z \|^2\\
\end{align*}

\begin{align*}
&\text{Appling Barbalats Lemma}\\
1)& V(\eta)\geq 0 \checkmark \\
2)& \dot{V} \leq -g(t)\; \text{where g(t) is U.C.} \\
&x \in \mathcal{L}_{\infty} \rightarrow z \in \mathcal{L}_{\infty} \Rightarrow g(t) \text{ is U.C.} \\
3)&g(t) \in \mathcal{L}_{2} \\
&g(t) \propto \left(byN(y) - \frac{\lambda^2}{2} - a\right)\|x\|^2 + \left(\frac{1}{2}\lambda_{min}(Q) - \frac{\beta}{2 \lambda^2} \right)\|z \|^2 \\
&g(t) = \| \eta \|^2\\
&\int_{0}^{T} \dot{V}(t) dt \leq -\int_{0}^{T} \| \eta \|^2 dt\\
&V(T) - V(0) \leq -\int_{0}^{T}\| \eta \|^2 dt\\
&\int_{0}^{T} \| \eta \|^2 dt \leq V(0) - V(T)\\ 
&\text{Due to 1}\rightarrow V(T) \geq 0\\
&\int_{0}^{T} \| \eta \|^2 dt \leq V(0) \\
&\int_{0}^{T} \| \eta \|^2dt \leq V(0) \leq \infty\\
& \text{Thus } \lim_{t\to\infty} \eta(t) = 0
\end{align*}
