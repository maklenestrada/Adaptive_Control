\begin{align*}
\text{Dynamics: }& M(q)\ddot{q} + V_{m}(q,\dot{q})\dot{q} + F_{s}(\dot{q})+ F_{d}(\dot{q}) + F_{k}(q) = \tau \\
\text{Angular Position (rad): }& q = [q_{1}, \; q_{2}]^{T}\\
\text{Angular Velocities (rad): }& \dot{q} = [\dot{q}_{1}, \; \dot{q}_{2}]^{T}
\end{align*}

\begin{align*}
\text{Tracking Error: }& e = q_{d} - q\\
\text{Filtered Tracking Error: }& r = \dot{e} + \alpha e \;\;\; \text{ where } \alpha > 0 \text{ is a design parameter} \\
\text{Error Dynamics: }& \dot{e} = r -\alpha e\\
& M(q)\dot{r} = Y(q,\dot{q},q_{d},dot{q}_d,\ddot{q}_d)\theta - V_{m}(q,\dot{q})r - \tau
\end{align*}
