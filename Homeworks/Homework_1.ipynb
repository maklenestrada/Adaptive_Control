#### Question 2

\begin{align*}
\text{Dynamics: }& M(q)\ddot{q} + V_{m}(q,\dot{q})\dot{q} + F_{s}(\dot{q})+ F_{d}(\dot{q}) + F_{k}(q) = \tau \\
\text{Angular Position (rad): }& q = [q_{1}, \; q_{2}]^{T}\\
\text{Angular Velocities (rad): }& \dot{q} = [\dot{q}_{1}, \; \dot{q}_{2}]^{T}
\end{align*}

\begin{align*}
\text{Tracking Error: }& e = q_{d} - q\\
\text{Filtered Tracking Error: }& r = \dot{e} + \alpha e \;\;\; \text{ where } \alpha > 0 \text{ is a design parameter} \\
\text{Error Dynamics: }& \dot{e} = r -\alpha e\\
& M(q)\dot{r} = Y(q,\dot{q},q_{d},\dot{q}_d,\ddot{q}_d)\theta - V_{m}(q,\dot{q})r - \tau
\end{align*}

\begin{align*}
\text{Linear regression equation: }& \tau_{f} = Y_{f}\theta\\
\text{Prediction error: }& \varepsilon = \tau_{f} - Y_{f}\hat{\theta} = Y_{f}\tilde{\theta} \\
\text{Design: }& \tau = Y\hat{\theta} + kr + e\;\;, \dot{\hat{\theta}}=\phi(t,r,\hat{\theta}),\;\;\;\;\; where\;\; k = k^{T} > 0\\
\end{align*}

\begin{align*}
\text{Plugging Controller in Dynamics : } M(q)\dot{r} &= Y\theta - V_{m}r - \left[Y\hat{\theta} + kr + e    \right] \\
 &= Y \tilde{\theta} - V_{m}r - kr - e
\end{align*}

$$ \text{Closed Loop: } \dot{z} = \begin{bmatrix} \dot{r} \\ \dot{e} \\ \dot{\tilde{\theta}} \end{bmatrix} = 
\begin{bmatrix} M^{-1}\left( Y \tilde{\theta} - V_{m}r - kr - e \right) \\  r -\alpha e \\ -\phi(t,r,\hat{\theta})      \end{bmatrix} =
f(t,z)
$$

\begin{align*}
\text{Candidate Lyapunov Function: } V(z)&= \frac{1}{2}r^{T}Mr + \frac{1}{2}e^{T}e + \frac{1}{2}\tilde{\theta}^{T}\Gamma^{-1}\tilde{\theta} \\

\text{Lie Derivative: } L_{f,t}V(t,z) &= 
r^{t}Y\tilde{\theta} - r^{t}V_{m}r- r^{t}kr -r^{T}e + \frac{1}{2}r^{T}\dot{M}r +
e^{T}r   - e^{T}\alpha e - 
\tilde{\theta}\Gamma^{-1}\phi(t,r,\hat{\theta})  
 \\

&= 
r^{t}Y\tilde{\theta} + r^{t}\left(\frac{1}{2}\dot{M} -   V_{m}\right)r- r^{T}e -  r^{t}kr  +
e^{T}r   - e^{T}\alpha e - 
\tilde{\theta}\Gamma^{-1}\phi(t,r,\hat{\theta})  \\

\text{Hint 3: }  x^{T}(\frac{1}{2}\dot{M} - V_{m})x &= 0 \\

&=
r^{t}Y\tilde{\theta} - r^{t}kr  -r^{T}e +
e^{T}r   - e^{T}\alpha e - 
\tilde{\theta}\Gamma^{-1}\phi(t,r,\hat{\theta})  \\

&=
r^{t}Y\tilde{\theta} - r^{t}kr  
- e^{T}\alpha e - 
\tilde{\theta}\Gamma^{-1}\phi(t,r,\hat{\theta})  \\

\text{Hint 4: If a matrix } S\in \mathbf{R}^{n\times n} \text{ is skew symmetric, then  } x^{T}Sx &= 0 \; \forall \; x\in\mathbf{R}^{n} \\

&=
r^{t}Y\tilde{\theta} -  r^{T}kr -
\tilde{\theta}\Gamma^{-1}\phi(t,r,\hat{\theta})  \\
\end{align*}

\begin{align*}
\text{Decision: } \dot{\hat{\theta}} &= \Gamma Y^{T}r + \Gamma Y_{f}^{T}\varepsilon \\

\rightarrow L_{f,t}V(t,z) &= 
r^{t}Y\tilde{\theta} -  r^{T}kr -
\tilde{\theta}\Gamma^{-1} \left( \Gamma Y^{T}r + \Gamma Y_{f}^{T}\varepsilon \right)\\

&= -  r^{T}kr - \tilde{\theta}Y_{f}^{T}\varepsilon \\

&= -  r^{T}kr - \tilde{\theta}Y_{f}^{T}Y_{f}\tilde{\theta}  \\

& \longrightarrow L_{f,t}V(t,z) = -  r^{T}kr - \tilde{\theta}Y_{f}^{T}Y_{f}\tilde{\theta}  \\
\end{align*}
