Show that the controller 
$$
\dot{y} = x^{2}, \;\; u = -yN(y)x
$$
where N(y) is a Nussbaum gain function, results in boundedness of all signals and convergence of the output x and the internal state z to zero regardless of the values of a, c, A, B, and b as long as $b \neq 0$

\begin{align*}
\text{Dynamics: } \dot{x} &= ax + bu + c^{T}z \\
\dot{z} &= Az + Bx\\
\text{Controller: } \dot{y} &= x^{2} \\
u &= -y N(y)x\\
\text{Closed Loop: } \dot{\eta} &= 
\begin{bmatrix} \dot{x} \\ \dot{z} \\ \dot{y}     \end{bmatrix}  \\
&= \begin{bmatrix} ax - byN(y)x + c^{T}z \\  Az + Bx  \\x^2    \end{bmatrix} \\
&= f(\eta)\\
\text{Lyapunov Function: } V(\eta) &= \frac{1}{2}x^2 + \frac{1}{2}z^{T}Pz\\
\text{Khalil Theorem 4.6 pg136 : } P = P^{T}>0&\text{ so that } A^{T}P+PA= -Q \text{ with } Q>0\\
\text{Derrivative of Lyapunov Function: } \frac{d}{dt}V(\eta) &= x\dot{x} + z^{T}P\dot{z} \\
&= x\left( ax - byN(y)x + c^{T}z \right) + z^{T}P\left(Az + Bx    \right)\\
&=  ax^{2} - byN(y)x^{2} + c^{T}zx+ z^{T}PAz + xz^{T}PB\\
&=  \left(a - byN(y)\right) \dot{y} + c^{T}zx+ z^{T}PAz + xz^{T}PB\\
&z^{T}PAz \text{   term}\\
z^{T}PAz &= (z^{T}PAz)^{T}\\
(z^{T}PAz)^{T} &= z^{T}A^{T}P^{T}z\\
&= z^{T}A^{T}Pz\\
\longrightarrow z^{T}PAz &= \frac{1}{2}\left( z^{T}PAz +  z^{T}A^{T}Pz  \right)\\
&= \frac{1}{2}z^{T}\left(PA + A^{T}P\right)z\\
&= -\frac{1}{2}z^{T} Q z\\
\frac{d}{dt}V(\eta) &= \left(a - byN(y)\right) \dot{y} + c^{T}zx-\frac{1}{2}z^{T} Q z+ xz^{T}PB\\
&= \left(a - byN(y)\right) \dot{y} -\frac{1}{2}z^{T}Qz + c^{T}zx + xz^{T}PB\\
&= \left(a - byN(y)\right) \dot{y} -\frac{1}{2}z^{T}Qz + x \left(c + PB\right)^{T}z
\end{align*}

\begin{align*}
\text{Completing the Square}\\
-\frac{1}{2}z^{T}Qz + x \left(c + PB\right)^{T}z
\end{align*}

### Need to work with $+ x \left(c^{T} + B^{T}P\right)z$ this term, d/dt of Lyapunov fxn not ND or NSD 

```python
#Import Libraries
import numpy as np 
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

#Hurwitz Matrix Generation 2x2
def HurwitzMatrix():
    R = np.random.randn(2,2)
    return -R.T @ R

#Nussbaum Gain Function
def NussbaumGain(y):
    return y*np.cos(y)

#Universal Regulator Dynamics
def UniversalRegulator(State,Parameters):
    #Extract State Variables
    x = State[0]
    z = State[1:3].reshape(2,1)
    y = State[3]
    
    #Extract Parameters
    a = Parameters.a
    b = Parameters.b
    c = Parameters.c
    A = Parameters.A
    B = Parameters.B

    #Controller 
    N_y = NussbaumGain(y)
    u = -y * N_y * x

    #System Dynamics
    x_dot = (a*x + b*u + c.T@z).item()
    z_dot = (A@z + B*u).flatten()
    y_dot = x**2

    return np.array([x_dot, z_dot[0], z_dot[1], y_dot])

#Universal Regulator Dynamics with Disturbance
def UniversalRegulator_Disturbance(State,Parameters):
    #Extract State Variables
    x = State[0]
    z = State[1:3].reshape(2,1)
    y = State[3]

    #Extract Parameters
    a = Parameters.a
    b = Parameters.b
    c = Parameters.c
    A = Parameters.A
    B = Parameters.B
    d = Parameters.d

    #Controller 
    N_y = NussbaumGain(y)
    u = -y * N_y * x

    #System Dynamics with Disturbance
    x_dot = (a*x + b*u + c.T@z + d).item()
    z_dot = (A@z + B*u).flatten()
    y_dot = x**2

    return np.array([x_dot, z_dot[0], z_dot[1], y_dot])
```

```python
#State Parameters
class Parameters:
    def __init__(self):
        self.a = np.random.randn()
        self.b = np.random.randn()
        self.c = np.random.randn(2,1)
        self.A = HurwitzMatrix()
        self.B = np.random.randn(2,1)
        self.d = 0.1
Param = Parameters()
Param_d = Parameters()
#Initial Conditions
x0 = 1
z0 = np.array([0, 0])
y0 = 0
InitialState = np.array([x0, z0[0], z0[1], y0])

#Simulation Time
tspan = (0, 500)
t_eval = np.linspace(tspan[0], tspan[1], num=100)

#Simulations
New_State = solve_ivp(lambda t, y: UniversalRegulator(y, Param), tspan, InitialState, t_eval=t_eval)   
New_State_Disturbance = solve_ivp(lambda t, y: UniversalRegulator_Disturbance(y, Param_d), tspan, InitialState, t_eval=t_eval)
```

```python
#Extract Results
t = New_State.t
x = New_State.y[0, :]
z1 = New_State.y[1, :]
z2 = New_State.y[2, :]
y_state = New_State.y[3, :]

t_d = New_State_Disturbance.t
x_d = New_State_Disturbance.y[0, :]
z1_d = New_State_Disturbance.y[1, :]
z2_d = New_State_Disturbance.y[2, :]
y_d = New_State_Disturbance.y[3, :]

# Plot without disturbance
plt.figure(figsize=(12, 8))
plt.subplot(4, 1, 1)
plt.title('System States vs Time - No Disturbance', fontsize=14)
plt.plot(t, x, 'b-', linewidth=1.5)
plt.ylabel('x', fontsize=12)
plt.grid(True)

plt.subplot(4, 1, 2)
plt.plot(t, z1, 'r-', linewidth=1.5)
plt.ylabel(r'$z_1$', fontsize=12)
plt.grid(True)

plt.subplot(4, 1, 3)
plt.plot(t, z2, 'g-', linewidth=1.5)
plt.ylabel(r'$z_2$', fontsize=12)
plt.grid(True)

plt.subplot(4, 1, 4)
plt.plot(t, y_state, 'k-', linewidth=1.5)
plt.ylabel('y', fontsize=12)
plt.xlabel('Time (s)', fontsize=12)
plt.grid(True)
plt.tight_layout()

# Plot with disturbance
plt.figure(figsize=(12, 8))
plt.subplot(4, 1, 1)
plt.title('System States vs Time - With Disturbance', fontsize=14)
plt.plot(t_d, x_d, 'b-', linewidth=1.5)
plt.ylabel('x', fontsize=12)
plt.grid(True)

plt.subplot(4, 1, 2)
plt.plot(t_d, z1_d, 'r-', linewidth=1.5)
plt.ylabel(r'$z_1$', fontsize=12)
plt.grid(True)

plt.subplot(4, 1, 3)
plt.plot(t_d, z2_d, 'g-', linewidth=1.5)
plt.ylabel(r'$z_2$', fontsize=12)
plt.grid(True)

plt.subplot(4, 1, 4)
plt.plot(t_d, y_d, 'k-', linewidth=1.5)
plt.ylabel('y', fontsize=12)
plt.xlabel('Time (s)', fontsize=12)
plt.grid(True)
plt.tight_layout()

plt.show()
```
